```python
import matplotlib.pyplot as mp
import numpy as np

%matplotlib inline
```

```python

q = int(1e5)
n = np.zeros(q)
a = 1.000001
n[0] = 1
for i in range(1,q): 
    n[i] = a*n[i-1]
mp.subplot(2,2,1)
mp.plot(n)

n = np.zeros(q)
a = 0.99999
n[0] = 1
for i in range(1,q): 
    n[i] = a*n[i-1]
   
mp.subplot(2,2,2)
mp.plot(n)

n = np.zeros(q)
a = 1
n[0] = 1
for i in range(1,q): 
    n[i] = a*n[i-1]

mp.subplot(2,2,3)
mp.plot(n)


```
